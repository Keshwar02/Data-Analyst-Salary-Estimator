### Analyst in Title

```python
cnt=0
for i in df['Job_Title']:
    if 'analyst' in i.lower():
        cnt += 1
print(cnt)
```

```python
def Job_Title(title):
    if 'data analyst' in title.lower():
        return 'Data Analyst'
    elif 'business analyst' in title.lower():
        return 'Business Analyst'
    elif 'financial analyst' in title.lower():
        return 'Financial Analyst'
    else:
        return 'NA'
```

```python
df['Job_title_simp'] = df['Job_Title'].apply(Job_Title)
```

```python
df['Job_title_simp'].value_counts()
```

```python
df['Job_Title'].value_counts()
```

### Seniority check

```python
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'Senior'
    elif 'jr' in title.lower():
        return 'Junior'
    elif 'fresher' in title.lower():
        return 'Fresher'
    elif 'inter' in title.lower():
        return 'Intern'
    else:
        return 'NA'
```

```python
df['seniority'] = df['Job_Title'].apply(seniority)
```

```python
df['seniority'].value_counts()
```
