```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import time

import pandas as pd
```

```python
# Initalize the WebDriver
driver = webdriver.Chrome()
driver.maximize_window()
driver.get('https://www.glassdoor.co.in/Job/india-data-analyst-jobs-SRCH_IL.0,5_IN115_KO6,18.htm')

# Initalize empty lists to store data
jobs = []
num_jobs_to_scrape = 2000
jobs_scraped = 0

# Login
try:
    first_btn = WebDriverWait(driver, 15).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[1]/div[2]/ul/li[2]/div/div/div[1]/div[1]')))
    first_btn.click()

    time.sleep(2)

    username_field = WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[7]/div[2]/div[2]/div[1]/div/div[2]/div/div/div/div/div/div/form/div[1]/div/div[1]/input')))
    username_field.send_keys('keshwarreddy02@gmail.com')

    countinue_btn_username = WebDriverWait(driver, 15).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[7]/div[2]/div[2]/div[1]/div/div[2]/div/div/div/div/div/div/form/div[2]/div/button')))
    countinue_btn_username.click()

    time.sleep(2)

    password_field = WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[7]/div[2]/div[2]/div[1]/div/div[2]/div/div/div/div/div/div/form/div[1]/div/div/div[1]/input')))
    password_field.send_keys('1234.Tharun')

    login_btn = WebDriverWait(driver, 15).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[7]/div[2]/div[2]/div[1]/div/div[2]/div/div/div/div/div/div/form/div[2]/div/button')))
    login_btn.click()

    time.sleep(2)

except Exception as e:
    print('An error occurred in the login phase')
    print(f'An error occurred: {e}')
finally:
    pass

while jobs_scraped < num_jobs_to_scrape:
    initial_listings = driver.find_elements(By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[1]/div[2]/ul/li[position()>last()-29]') # Select the last 30 new listings
    num_initial_listings = len(initial_listings)
    print(f"Found {num_initial_listings} new listings on this page.")

    for job_index in range(1, num_initial_listings + 1):
        if jobs_scraped >= num_jobs_to_scrape:
            break
        try:
            job_card_xpath = f'/html/body/div[1]/div[5]/div[4]/div[2]/div[1]/div[2]/ul/li[last()-{num_initial_listings - job_index}]/div/div'
            job_card = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, job_card_xpath)))
            job_card.click()

            time.sleep(3)

            # Fetching Basic Job Details
            try:
                job_title = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/header/div[1]/div[1]/h1'))).text
                time.sleep(0.2)
            except NoSuchElementException:
                job_title = -1

            try:
                employer = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/header/div[1]/div[1]/a/div[2]/div/div[1]/h4'))).text
                time.sleep(0.2)
            except NoSuchElementException:
                employer = -1

            try:
                rating = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/header/div[1]/div[1]/a/div[2]/div/div[2]/span'))).text
                time.sleep(0.2)
            except NoSuchElementException:
                rating = -1

            try:
                location = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/header/div[2]/div[1]'))).text
                time.sleep(0.2)
            except NoSuchElementException:
                location = -1

            try:
                salary = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/header/div[2]/div[3]'))).text
                time.sleep(0.2)
            except NoSuchElementException:
                salary = -1

            print(job_title)
            print(employer)
            print(rating)
            print(location)
            print(salary)

            # Fetching Company Overview Details
            time.sleep(4)
            overview_section = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/section/div[4]/section/div/h2/span')))
            driver.execute_script('arguments[0].scrollIntoView();',overview_section)
            time.sleep(2)
            try:
                try:
                    size = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/section/div[4]/section/div/div/div[1]/div'))).text
                    time.sleep(0.2)
                except NoSuchElementException:
                    size = -1

                try:
                    founded_year = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/section/div[4]/section/div/div/div[2]/div'))).text
                    time.sleep(0.2)
                except NoSuchElementException:
                    founded_year = -1

                try:
                    type = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/section/div[4]/section/div/div/div[3]/div'))).text
                    time.sleep(0.2)
                except NoSuchElementException:
                    type = -1
                try:
                    industry = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/section/div[4]/section/div/div/div[4]/div'))).text
                    time.sleep(0.2)
                except NoSuchElementException:
                    industry = -1

                try:
                    sector = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/section/div[4]/section/div/div/div[5]/div'))).text
                    time.sleep(0.2)
                except NoSuchElementException:
                    sector = -1

                try:
                    revenue = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[2]/div/div[1]/section/div[4]/section/div/div/div[6]/div'))).text
                    time.sleep(0.2)
                except NoSuchElementException:
                    revenue = -1

            except NoSuchElementException:
                size = -1
                founded_year = -1
                type = -1
                industry = -1
                sector = -1
                revenue = -1

            # Printing for debugging
            print(size)
            print(founded_year)
            print(type)
            print(industry)
            print(sector)
            print(revenue)

            jobs.append({'Job_Title': job_title,
                         'Company_Name': employer,
                         'Ratings': rating,
                         'Location': location,
                         'Salary_Estimates': salary,
                         'Company_Size': size,
                         'Founding_Year': founded_year,
                         'Type': type,
                         'Industry': industry,
                         'Sector': sector,
                         'Revenue': revenue})
            
            jobs_scraped += 1
            print(f"Scraped {jobs_scraped} jobs so far.")

        except Exception as e:
            print('Error in 2nd Try & Catch block')

    if jobs_scraped < num_jobs_to_scrape:
        print("Clicking 'Show More Jobs'...")
        try:
            show_more_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[5]/div[4]/div[2]/div[1]/div[2]/div/div/button'))
                                                       
            )
            show_more_button.click()
            time.sleep(5)  # Give time for new jobs to load
        except TimeoutException:
            print("Could not find or click 'Show More Jobs' button. Stopping.")
            break # Stop if the button isn't found
        except Exception as e:
            print(f"An error occurred while clicking 'Show More Jobs': {e}. Stopping.")
            break # Stop if there's an error clicking

print(f"Successfully scraped {len(jobs)} job listings.")
# You can now process the 'jobs' list
# driver.quit()
```
